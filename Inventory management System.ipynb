```python
def menuDisplay(): ##displaying the various option you can go for
    print('==============================================================================================================')
    print('===============================Inventory Management System ===========================================')
    print('==============================================================================================================')
    print('(1) Add New Item')
    print('(2) Remove Item')
    print('(3) Purchasing & Update Item')
    print('(4) Search Item')
    print('(5) Print Inventory Report')
    print('(6) Quit')
    print('--------------------------------------------------------------------------------------------------------------')
    CHOICE = int(input("Enter choice: "))
    menuSelection(CHOICE)

def menuSelection(CHOICE): ##choose the operation you want to perform
    if CHOICE == 1:
        addInventory()
    elif CHOICE == 2:
        removeInventory()
    elif CHOICE == 3:
        purchasing()
    elif CHOICE == 4:
        searchInventory()
    elif CHOICE == 5:
        printSaleList()
    elif CHOICE == 6:
        return
    else:
        print("Sry wrong input! try again")

#ADDING NEW PRODUCT IN THE INVENTORY
def addInventory():
    
    print("===================================Adding Inventory======================================")
    print("=======================================================================================")
    product_id = input("Enter the product_id of the item: ")
    name=input("enter the name: ")
    price=input("enter the price: ")
    quantity_in_stock=input("enter the quantity: ")
    rating=input("enter the rating: ")
    type_of_product=input("enter the type_of_product: ")
    records[product_id]={"name": name, "price": price, "quantity_in_stock": quantity_in_stock, "rating": rating, "type_of_product": type_of_product}
    
    # to save the products in the file for futher use
    js = json.dumps(records)

    fd = open("records.json",'a') #'a' use to STORE THE NEW ADDED PRODUCTED TO THE EXISTING LIST
    fd.write(js)
    fd.close()
    
    CHOICE = input('Enter y to continue or n to exit: ') 
    if CHOICE == 'y':
        menuDisplay()
    else:
        return
 #remove product for the records.json    
def removeInventory():
    print("=========================================Removing Inventory============================================")
    print("=======================================================================================================")
    product_id = input("Enter the item name to remove from inventory: ")
    records.pop(product_id)
    print("Product has been deleted successfully....\n")
    #to store the change in the file
    js = json.dumps(records)

    fd = open("records.json",'w')
    fd.write(js)
    fd.close()
    CHOICE = input('Enter y to continue or n to exit: ')
    if CHOICE == 'y':
        menuDisplay()
    else:
        return
#purchasing
def purchasing():
    print("==========================================Purchasing====================================================")
    print("=======================================================================================================")
    
    ui_prod  = str(input("Enter the product_Id: "))
    ui_quant = int(input("Enter the quantity: "))
    if(ui_quant >= records[ui_prod]['quantity_in_stock']):# if the quantity asked by the buyers is less than the quantity present 
        print("OUT OF STOCK")
        print("We only have ", records[ui_prod]['quantity_in_stock'])
        
    else: # if the quantity of stock is sufficient 
        print("   $$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$-----Bill-----$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$$    ")
        print("-----------------------------------------------------------------------------------------------------------")
       
        print("Product: ", records[ui_prod]['name'])
        print("Quantity Purchased", ui_quant)
        print("Price: ", records[ui_prod]['price'])
        print("Billing Amount: ", records[ui_prod]['price'] * ui_quant) # to get the total amount buyers need to pay

        records[ui_prod]['quantity_in_stock'] = records[ui_prod]['quantity_in_stock'] - ui_quant #the quantity left after selling
        
        js = json.dumps(records)

        fd = open("records.json",'w')
        fd.write(js)
        fd.close()
        
        #Saleslist 
        sales={ "Prod_id":ui_prod,"Prod_name":records[str(ui_prod)]['name'], "quantity":ui_quant, "amount":records[ui_prod]['price'] * ui_quant}
        fs = open("sales.json",'a')
        sd=json.dumps(sales)
        fs.write(sd)
        fs.close()                                   
                
    CHOICE = input('Enter y to continue or n to exit: ')
    if CHOICE == 'y':
        menuDisplay()
    else:
        return
# for searching of product required
def searchInventory():
    print('===================================================Searching Inventory=====================================')
    print('===========================================================================================================')
    product_id = input('Enter the name of the item: ')
    product_name=input('Enter the name of the product: ')
    f = open('records.json', 'r')
    search = f.readlines()
    f.close
   
    if product_id in records.keys() or product_name in records.values:
        print("product is available")
    else:
        print("product does not exist")
        
    CHOICE = input('Enter y to continue or n to exit: ')
    if CHOICE == 'y':
            menuDisplay()
    else:
        return
# to display the sales list of the shop       
def printSaleList():
                      
    
    print("***************************************** SALES LIST ********************************************\n")
    fdr = open("sales.json" , 'r')
    txt1 = fdr.read()
    fdr.close()
    print(txt1)
    CHOICE = input('Enter y to continue or n to exit: ')
    if CHOICE == 'y':
        menuDisplay()
    else:
        return

menuDisplay()
```
